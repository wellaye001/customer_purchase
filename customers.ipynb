### How should we handle missing values? (Should we drop the rows? or impute them with values?)

```python
#Find which columns have missing value
data.columns[data.isnull().any()]
```

#### Handling Missing Value for `Cust_Last_Purchase` column

```python
#Impute the missing values for Cust_Last_Purchase column by median because of very high outlier (200)
data.Cust_Last_Purchase.isnull().sum()
```

```python
data['Cust_Last_Purchase'] = data['Cust_Last_Purchase'].fillna(0)
```

#### Handling Missing Value for `Pur_3_years_Avg_Indirect` column 

```python
#Impute the missing values for Pur_3_years_Avg_Indirect column
data.Pur_3_years_Avg_Indirect.describe()
```

```python
display(data.Pur_3_years_Avg_Indirect.isnull().sum())

nullcol = ["Pur_3_years_Avg_Indirect"]
for i in nullcol:
    data.loc[data.loc[:, i].isnull(),i] = data.loc[:,i].median()
    
data.Pur_3_years_Avg_Indirect.isnull().sum()
```

#### Handling Missing Value for `Age` column

Changes for Age is omitted since it is not relevant to change missing value of age to 0.

```python
data.isnull().sum()
```

```python
#Impute the missing values for Age column
data.Age.describe()
```

```python
#Replace the missing values(values lower than 14) for Age column 
nullcol = ["Age"]
for i in nullcol:
    data.loc[data.loc[:, i].isnull(),i] = data.loc[:,i].mean()
    data.loc[data.loc[:, i] <= 14,i] = data.loc[:,i].median()
```

#### Handling Missing Value for `Cust_Ann_Income` column 

```python
data.Cust_Ann_Income.isnull().any()
zeroann= ["Cust_Ann_Income"]
for i in zeroann:
    data.loc[data.loc[:, i] == 0,i] = data.loc[:,i].mean()
```

#### Handling Missing Value for `Gender` column

```python
#Replace the missing values for Gender column
data['Gender'] = np.where(data.Gender == 'U', data.Gender.mode(), data.Gender)
data['Gender'].unique()
```
